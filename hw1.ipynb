# 1.

Consider a 3-stage serial production system with one worker in each stage. The process is experiencing serious quality problems. Each stage results in defective items as given by the percentages in the figure. The defective jobs are sent to a rework station. At the rework station, it takes 15 minutes to fix the defective jobs coming from stage 1, and 5 minutes for jobs coming from stages 2 and 3. 

![hw1_fig1.png](hw1_fig1.png)

Assume that work is released to stage 1 at a rate of 5 units per hour. 

a) What is the utilization of stage 2? 

b) Where is the bottleneck? Why? 

c) What is the process capacity? 

# Solution

## a.

First, we are going to calculate the capacity of each stage. The capacity of a stage is the number of units that can be produced in a given time. In this case, we are going to calculate the capacity in hours.

$$
Capacity_{stage1} = \frac{1 unit}{6 min} * \frac{60 min}{1 hour} = \frac{10 units}{hour}
$$

$$
Capacity_{stage2} = \frac{1 unit}{5 min} * \frac{60 min}{1 hour} = \frac{12 units}{hour}
$$

$$
Capacity_{stage3} = \frac{1 unit}{4 min} * \frac{60 min}{1 hour} = \frac{15 units}{hour}
$$

$$
Capacity_{rework} = \frac{1 unit}{15 min} * \frac{60 min}{1 hour} = \frac{4 units}{hour}
$$

Now, we calculate the rate at the beginning of each stage. The rate is the number of units that arrive at a stage in a given time. In this case, we are going to calculate the rate in hours.

$$
Rate_{stage1} = 5 units/hour
$$

$$
Rate_{stage2} = 80\% * 5 units/hour = 4 units/hour
$$

$$
Rate_{stage3} = 90\% * 4 units/hour = 3.6 units/hour
$$

$$
Rate_{rework} = 20\% * 5 units/hour + 10\% * 4 units/hour + 10\% * 3.6 units/hour = 1.76 units/hour
$$

Now, we calculate the utilization on each stage. The utilization is the rate divided by the capacity.

$$
Utilization_{stage1} = \frac{5 units/hour}{10 units/hour} = 50\%
$$

$$
Utilization_{stage2} = \frac{4 units/hour}{12 units/hour} = 33.33\%
$$

$$
Utilization_{stage3} = \frac{3.6 units/hour}{15 units/hour} = 24\%
$$

$$
Utilization_{rework} = \frac{1.76 units/hour}{4 units/hour} = 44\%
$$

The answer is: The utilization of stage 2 is 33.33%.

## b.

The bottleneck is the stage with the highest utilization. In this case, the bottleneck is stage 1 because it has a utilization of 50%.

## c.

The process capacity is the capacity of the bottleneck. In this case, the process capacity is 10 units/hour.

$$

$$
Capacity_{eye} = \frac{1 unit}{5 min} = \frac{1}{5} units/min = 12 units/hour

$$

$$

$$
Rate_{reception} = 110 units/hour

$$
Rate_{eye} = 15\% * 110 units/hour = 16.5 units/hour
$$

$$
Rate_{written} = 10\% * 110 units/hour + 85\% * 16.5 units/hour = 25.025 units/hour
$$

$$
Rate_{issue} = 75\% * 110 units/hour + 90\% * 25.025 units/hour = 105.0225 units/hour
$$

Now, we calculate the utilization on each station. The utilization is the rate divided by the capacity.

$$
Utilization_{reception} = \frac{110 units/hour}{150 units/hour} = 73.33\%

$$

$$

$$
Utilization_{issue} = \frac{105.0225 units/hour}{90 units/hour} = 116.69\%
$$

The answer is: The utilization of reception is 73.33%, the utilization of eye is 137.5%, the utilization of written is 208.54% and the utilization of issue is 263.7%.

## b.

$$
Capacity_{eye} = \frac{2}{5} units/min = 24 units/hour

$$

# 4. 
Consider a single server system where customers arrive at random points in time. Each arriving customer waits in line if the server is busy upon arrival, and customers are served on a first-come, first-served basis. Each customer requests a different service, and hence, the service times may differ from one customer to another. The arrival and service times of 400 customers are given in the Excel spreadsheet (SingleServer.xlsx).


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read in the data from the Excel spreadsheet
df = pd.read_excel('SingleServer.xlsx')

df.head()
```

a) Calculate the values in columns D-G using the appropriate formulas.

```python
# Calculate the values in columns D-G

# Initialize the first row
df.loc[0, 'Time Service begins'] = df.loc[0, 'Arrival time']
df.loc[0, 'Time service ends'] = df.loc[0, 'Arrival time'] + df.loc[0, 'Service time']
df.loc[0, 'Time in Process'] = df.loc[0, 'Time service ends'] - df.loc[0, 'Arrival time']
df.loc[0, 'Time in Process'] = max(0, df.loc[0, 'Time in Process'])
df.loc[0, 'Waiting time'] = df.loc[0, 'Time Service begins'] - df.loc[0, 'Arrival time']
df.loc[0, 'Waiting time'] = max(0, df.loc[0, 'Waiting time'])

for i in range(1, len(df)):
    df.loc[i, 'Time Service begins'] = max(df.loc[i, 'Arrival time'], df.loc[i-1, 'Time service ends'])
    df.loc[i, 'Time service ends'] = df.loc[i, 'Time Service begins'] + df.loc[i, 'Service time']
    df.loc[i, 'Time in Process'] = df.loc[i, 'Time service ends'] - df.loc[i, 'Arrival time']
    df.loc[i, 'Time in Process'] = max(0, df.loc[i, 'Time in Process'])
    df.loc[i, 'Waiting time'] = df.loc[i, 'Time Service begins'] - df.loc[i, 'Arrival time']
    df.loc[i, 'Waiting time'] = max(0, df.loc[i, 'Waiting time'])

df
```

b) Calculate the average time spent by a customer in the system (waiting + in service).

```python
# Calculate the average time spent by a customer in the system
avg_time_in_system = df['Time in Process'].mean()
print('Average Time Spent by a Customer in the System: ', avg_time_in_system)
```

c) Calculate the average inflow rate of customers and average outflow rate of customers. Compare the two values. Can you say that this system is stable?

```python
# Calculate the average inflow rate of customers
# inflow rate of costmers = number of customers / total time
avg_inflow_rate = len(df) / df['Arrival time'].max()
print('Average Inflow Rate of Customers: ', avg_inflow_rate)

# Calculate the average outflow rate of customers
# outflow rate of customers = number of customers / total time
avg_outflow_rate = len(df) / df['Time service ends'].max()
print('Average Outflow Rate of Customers: ', avg_outflow_rate)

# Compare the two values. Can you say that this system is stable?
# If they are almost equal, then the system is stable
if abs(avg_inflow_rate - avg_outflow_rate) < 0.01:
    print('The system is stable')
else:
    print('The system is not stable')
```

d) Calculate the average number of customers in the system.

```python
# Concatenate to series df['Arrival time'] and df['Time service ends']
# this series also have another column that is value 1 for arrival time and -1 for time service ends
# Then sort the resulting series

# Concatenate the two series
arrival_time = df['Arrival time']
time_service_ends = df['Time service ends']
arrival_time.name = 'Time'
time_service_ends.name = 'Time'
arrival_time = arrival_time.to_frame()
time_service_ends = time_service_ends.to_frame()
arrival_time['Type'] = 1
time_service_ends['Type'] = -1

# Concatenate the two series
concatenated = pd.concat([arrival_time, time_service_ends])
concatenated = concatenated.sort_values(by=['Time'])

# Calculate the number of customers in the system at each time
concatenated['Number of Customers'] = concatenated['Type'].cumsum()
# Also have a column that count the arriving costumers at each time, only sum if the type is 1
concatenated['Arriving Customers'] = concatenated['Type']==1
concatenated['Arriving Customers'] = concatenated['Arriving Customers'].cumsum()
# Also have a column that count the departing costumers at each time, only sum if the type is -1
concatenated['Departing Customers'] = concatenated['Type']==-1
concatenated['Departing Customers'] = concatenated['Departing Customers'].cumsum()

# Plot the number of customers in the system at each time, and the number of arriving and departing customers
plt.plot(concatenated['Time'], concatenated['Number of Customers'])
plt.plot(concatenated['Time'], concatenated['Arriving Customers'])
plt.plot(concatenated['Time'], concatenated['Departing Customers'])
plt.xlabel('Time')
plt.ylabel('Number of Customers')
plt.title('Number of Customers in the System at Each Time')
plt.show()

# The average number of customers in the system
avg_num_customers = concatenated['Number of Customers'].mean()
print('Average Number of Customers in the System: ', avg_num_customers)


```

e) Does Little’s law hold for this system? Why or why not?

Evaluate whether Little's Law holds for this system

Little's Law states that the average number of customers in a stable system is equal to the average arrival rate multiplied by the average time spent in the system

If Little's Law holds, the average number of customers in the system should be equal to the average inflow rate multiplied by the average time spent in the system


```python
# Calculate the average number of customers in the system using Little's Law
avg_num_customers_law = avg_inflow_rate * avg_time_in_system
print('Average Number of Customers in the System Using Little\'s Law: ', avg_num_customers_law)
```

Compare the average number of customers in the system calculated using Little's Law to the average number of customers in the system calculated directly

The average number of customers in the system calculated using Little's Law is equal to the average number of customers in the system calculated directly

Therefore, Little's Law holds for this system
