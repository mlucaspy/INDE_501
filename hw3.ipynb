# 1.
The buyer for Needless Markup, a famous ”high end” department store, must decide on the quantity of a high-priced woman’s handbag to procure from Italy for the following Christmas season. The unit cost of the handbag to the store is $28.5, and the handbag will sell for $150. Any handbags not sold by the end of the season are purchased by a discount firm for $20. In addition, the store accountants estimate that there is a cost of $0.4 for each dollar tied up in inventory, as this dollar invested elsewhere could have yielded a gross profit. Assume that this cost is attached to unsold bags only.

a) Suppose that the sales of the bags are equally likely to be anywhere from 50 to 250 handbags during the season. Based on this, how many bags should the buyer purchase?

b) A detailed analysis of past data shows that the number of bags sold is better described by a normal distribution, with mean 150 and standard deviation 20. Now what is the optimal number of bags to be purchased?

c) The expected demand was the same in parts (a) and (b), but the optimal order quantities should have been different. What accounted for this difference?

# Solution

<!-- use the newsvendor model -->

$C_o$: cost per unit left over after demand is realized.

$$
C_o = 28.5 - 20 + 0.4(28.5) = 19.9
$$

$C_u$: cost per unit of demand that is not satisfied.

$$
C_u = 150 - 28.5 = 121.5
$$

The critical ratio is

$$
\frac{C_o}{C_o + C_u} = \frac{19.9}{19.9 + 121.5} \approx 0.1407
$$

### a.

The probability function of the Demand is uniform from 50 to 250.

$$
\begin{cases}
p(x) = \frac{1}{200} & 50 \leq x \leq 250 \\
p(x) = 0 & \text{otherwise}
\end{cases}
$$

The formula for the optimal solution is

$$
F(Q^*) = P(D \leq Q^*) = \frac{C_o}{C_o + C_u}
$$

The formula for the optimal order quantity is

$$
Q^* = F^{-1}\left(\frac{C_o}{C_o + C_u}\right)
$$

$$
Q^* = F^{-1}(0.1407) = 50 + 200(0.1407) = 78.14
$$

### b.

Now the probability function of the Demand is normal with mean 150 and standard deviation 20.

$$
p(x) = \frac{1}{20\sqrt{2\pi}}e^{-\frac{(x-150)^2}{2(20)^2}}
$$

TThe optimal solution is

$$
Q^* = F^{-1}\left(\frac{C_o}{C_o + C_u}\right) = F^{-1}(0.1407)
$$

For a cumulative probabilty of 0.1407, the $Z$-score is -1.0771. So

$$
Z = \frac{Q^* - 150}{20} = -1.0771
$$

Then the optimal order quantity is

$$
Q^* = 150 + 20(-1.0771) = 128.46
$$

### c.

The Expected value of the Demand is the same in both cases, but the standard deviation is different. The standard deviation is a measure of the uncertainty of the demand. The higher the standard deviation, the higher the uncertainty. The higher the uncertainty, the higher the optimal order quantity.

One of their popular products for the Halloween season is the Harry Potter (HP) costume for kids, complete with robe, glasses frame and the wizard wand. The HP costume is produced in China, so P-City must make a single order well in advance of the upcoming season. The management of P-City expects demand to be 25,000 units and the following is the demand distribution based on the demand forecasts.

# Solution

### a.

If the order is the 40,000 and 10,000 or more have to be sold to the discount store. The demand should be $D \le 30,000$. 

So, $P(D \le 30,000)$ is $F(25000) = 0.6289$

### b.

To find the optimal order quantity, we can use the Newsvendor model. The critical ratio $ F(Q^*) $ is calculated as follows:

$$
F(Q^*) = \frac{C_o}{C_o + C_u}
$$

Where $ C_o $ is the overage cost, i.e., the cost of ordering one too many (Cost - Salvage value) and $ C_u $ is the underage cost, i.e., the cost of ordering one too few (Selling price - Cost).

- $ C_o = 20 - 12 = 8 $
- $ C_u = 45 - 20 = 25 $

So,

$$
F(Q^*) = \frac{8}{8 + 25} = \frac{8}{33} \approx 0.2424
$$

We look for the quantity \( Q \) such that \( F(Q) \) is closest to 0.2424. From the table, $ F(15,000) = 0.2381 $ is the closest. Therefore, the optimal order quantity is 15,000 units.

### c.

If P-City wants to have at least a 90% service level, so it means that the probability of demand is less than or equal to the order quantity is 90%. So, we need to find the quantity \( Q \) such that \( F(Q) \) is closest to 0.9. From the table, $ F(35,000) = 0.8894 $ is the closest. Therefore, the optimal order quantity is 35,000 units.

### d.

### 100% In-Stock Probability and Expected Profit

To have a 100% in-stock probability, P-City would have to order for the maximum possible demand, which is 75,000 according to your table.

The expected profit can be calculated as follows:

$$
Expected Profit = (SellingPrice - Cost) \times ExpectedDemand - (Cost - SalvageValue) \times (MaximumDemand - ExpectedDemand)
$$
$$
= (45 - 20) \times 25,000 - (20 - 12) \times (75,000 - 25,000)
$$
$$
= 25 \times 25,000 - 8 \times 50,000
$$
$$
= 625,000 - 400,000 = 225,000
$$

So, the expected profit would be $225,000 if P-City insists on a 100% in-stock probability.





# 3.

An automotive warehouse stocks a variety of parts that are sold at neighborhood stores. One particular part, a popular brand of oil filter, is purchased by the warehouse for $1.5 each. It is estimated that the cost of order processing and receipt is $100 per order. The company uses an inventory carrying charge based on a 28 percent annual interest rate.
The monthly demand for the filter follows a normal distribution with mean 280 and standard deviation 77. Order lead time is assumed to be five months. Assume that the management would like to achieve a service level of 95%.

a)	Determine the order quantity and the reorder point. What is the safety stock?

b)	How will your results change if the service level is increased to 99%? Please calculate.


# Solution

The monthly demand for the filter follows a normal distribution with mean 280 and standard deviation 77. Order lead time is assumed to be five months. Assume that the management would like to achieve a service level of 95%.

a)	Determine the order quantity and the reorder point. What is the safety stock?

### a.

The formula for the optimal order quantity for a service level of 95% is

$$
F(Q^*) = \frac{C_o}{C_o + C_u} = 0.95
$$

The formula for the optimal order quantity given a normal distribution is

$$
P(D \leq Q^*) = 0.95 \\
P(Z \leq \frac{Q^* - \mu}{\sigma}) = 0.95 \\
\frac{Q^* - \mu}{\sigma} = Z_{0.95} \\
Q^* = \mu + Z_{0.95}\sigma
$$

The order quantity is

$$
Q^* = 280 + 1.645(77) = 402.465
$$

The formula for the reorder point is

$$
R = \mu + Z_{0.95}\sigma + L
$$

Where $ L $ is the lead time. The reorder point assuming a lead time of 5 months is

$$
R = 280 + 1.645(77) + 5 = 407.465
$$

The safety stock is

$$
S = R - \mu = 407.465 - 280 = 127.465
$$

### b.

The formula for the optimal order quantity for a service level of 99% is

$$
F(Q^*) = \frac{C_o}{C_o + C_u} = 0.99
$$

The order quantity is

$$
Q^* = 280 + 2.326(77) = 436.402
$$

The safety stock is

$$
S = R - \mu = 436.402 - 280 = 156.402
$$
