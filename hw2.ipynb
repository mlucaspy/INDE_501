# 1.
DowChem, a manufacturer specializing in chemical products, currently serves 20 customers within the same region through its distribution center. These customers can be categorized into three types: small, medium, and large, each with their respective annual consumption, as detailed in the following table:

| Customer Type | Number of Customers | Consumption (Pounds per year) |
|---------------|---------------------|-----------------------------|
| Small         | 12                  | 12,000                      |
| Medium        | 6                   | 60,000                      |
| Large         | 2                   | 144,000                     |

DowChem collaborates with Superior Trucking Co. for all its shipping needs. Each truck employed for transportation has a maximum capacity of 40,000 pounds. Superior charges a fixed fee of $400 for each shipment, irrespective of the quantity transported on the truck—from the distribution center to each customer.

DowChem's current approach is to dispatch a full truckload to each customer as required. Nevertheless, John Smith, the Vice President of DowChem, is deeply concerned about the escalating distribution costs. John approached Superior to inquire about the possibility of consolidating shipments for multiple customers onto a single truck. Superior conveyed that they would charge $350 per truck and an additional $50 for each drop-off they manage. Consequently, if a single truck is used for a solitary delivery, the total charge would amount to $400. However, if a truck has to make four deliveries, the total charge would rise to $550.

The cost incurred by DowChem for each pound of chemical is $1, and the company also bears a holding cost of 25%. John is eager to evaluate several distribution options to determine an improved strategy:

**Option 1:** Consolidating all 20 customers into one truck. Whenever a truck is dispatched, it will serve all customers.

**Option 2:** Grouping the 20 customers into two categories, with one large, three medium, and six small customers in each group. The shipment for each group will be aggregated into a single truck.

**Option 3:** Aggregating each customer type into one truck. This means that large customers are bundled together, and their shipment is consolidated into a single truck, and the same applies to medium and small customers.

**a) (10 pts)** What is the annual cost associated with DowChem's current strategy of delivering full truckloads to each customer to replenish inventory?

**b) (20 pts)** Assuming the computation of the optimal shipment quantity using the EOQ (Economic Order Quantity) model and the optimal annual total cost for each of the delivery options 1, 2, and 3, which delivery option would you recommend for DowChem, and why?


# Solution

### Part a)

To calculate the annual cost of DowChem's current strategy, we need to consider the following costs:

1. **Shipping Cost:** This includes the fixed cost of USD400 per shipment.

2. **Holding Cost:** DowChem incurs a holding cost of 25% for each pound of chemical. We'll calculate this as a percentage of the annual consumption for each customer type and then sum them up.

3. **Total Cost:** The sum of the shipping cost and the holding cost.

Let's calculate these costs:

1. Shipping Cost:

   DowChem has 20 customers, and each customer is served individually. Therefore, the total annual shipping cost can be calculated as follows:

   For the small customers:

   - There would be no full truckload shipments for these customers since their annual consumption is less than the truck's capacity. Therefore, each shipment would cost USD400.
   - As they are 12 in number, the total cost would be USD400 x 12 = USD4,800.

   For the medium customers:
   - It would be necessary to dispatch at least two truckloads for these customers since their annual consumption is more than the truck's capacity. Therefore, each shipment would cost USD400.
   - Each costumet would incur a cost of USD400 x 2 = USD800.
   - As they are 6 in number, the total cost would be USD800 x 6 = USD4,800.

   For the large customers:
   - It would be necessary to dispatch at least four truckloads for these customers since their annual consumption is more than the truck's capacity. Therefore, each shipment would cost USD400.
   - Each costumet would incur a cost of USD400 x 4 = USD1,600.
   - As they are 2 in number, the total cost would be USD1,600 x 2 = USD3,200.

   Therefore, the total shipping cost would be:

   $$
   \text{Total Shipping Cost} = USD4,800 + USD4,800 + USD3,200 = USD12,800
   $$

2. Holding Cost:

   For each customer type (small, medium, and large), we calculate the annual holding cost based on their respective consumption. Then, we sum these costs.

   - For small customers, as they are 12 in number, the total annual consumption would be 12,000 x 12 = 144,000 pounds. Therefore, the annual holding cost would be 25% of 144,000 x USD1 = USD36,000.
   - For medium customers, as they are 6 in number, the total annual consumption would be 60,000 x 6 = 360,000 pounds. Therefore, the annual holding cost would be 25% of 360,000 x USD1 = USD90,000.
   - For large customers, as they are 2 in number, the total annual consumption would be 144,000 x 2 = 288,000 pounds. Therefore, the annual holding cost would be 25% of 288,000 x USD1 = USD72,000.

   Therefore, the total holding cost would be:
   
   $$
   \text{Total Holding Cost} = USD36,000 + USD90,000 + USD72,000 = USD198,000
   $$

3. Total Cost:

   Now, we sum up the shipping cost and the holding cost to get the total annual cost:

   $$
   \text{Total Annual Cost} = \text{Total Shipping Cost} + \text{Total Holding Cost}
   $$

   $$
   \text{Total Annual Cost} = USD12,800 + USD198,000 = USD210,800
   $$

Therefore, the annual cost of DowChem's current strategy is USD210,800.


# Part b)

Let's delve deeper into each of the options for DowChem's distribution strategy.

### Option 1: Consolidating All Customers into One Truck

**Delivery Process:**

1. **Preparation:** All orders from the 20 customers are accumulated until there's enough demand to fill a truck.
2. **Loading:** The truck is loaded up to its maximum capacity of 40,000 pounds. 
3. **Route Planning:** A delivery route is mapped out to service all 20 customers in the most efficient manner possible, ensuring that the travel time and distance are minimized.
4. **Delivery:** The truck starts from the distribution center and makes its way to each of the 20 customers. At each stop, the precise amount of chemical needed by that customer is offloaded. This means the truck will make 20 different stops during its route.
5. **Billing:** Superior Trucking Co. will bill DowChem $350 for the truck and an additional $50 for each of the 20 drop-offs. This amounts to a total of $1,350 for the entire route.

### Option 2: Grouping Customers into Two Categories

**Delivery Process:**

1. **Group Formation:** The 20 customers are divided into two groups. Each group consists of one large, three medium, and six small customers.
2. **Preparation:** Orders from customers in each group are collected separately. When there's enough demand within a group to fill a truck, that group's truck is dispatched.
3. **Loading:** Two trucks are loaded, each up to a maximum capacity of 40,000 pounds, based on the orders from their respective groups.
4. **Route Planning:** For each truck, a delivery route is designed to service the 10 customers in its group efficiently.
5. **Delivery:** Each truck starts from the distribution center and delivers to the 10 customers in its group. Every truck makes 10 stops.
6. **Billing:** For each truck, Superior Trucking Co. will bill DowChem $350 plus $50 for each of the 10 drop-offs, totaling $850. Since there are two trucks, the combined cost for both routes will be $1,700.

### Option 3: Aggregating Each Customer Type into One Truck

**Delivery Process:**

1. **Grouping by Type:** Customers are grouped based on their types—small, medium, and large.
2. **Preparation:** Orders are collected from all customers of a specific type until there's enough demand to fill a truck for that type.
3. **Loading:** Three separate trucks are loaded: one for small customers, one for medium customers, and one for large customers. Each truck is loaded up to its capacity based on the total orders from its specific customer type.
4. **Route Planning:** Each truck has a delivery route mapped out to service customers of its type.
5. **Delivery:** 
    - The truck for small customers will make 12 stops.
    - The truck for medium customers will make 6 stops.
    - The truck for large customers will make 2 stops.
6. **Billing:** 
    - For the small customer truck, Superior Trucking Co. will bill $350 plus $50 x 12 = $600, totaling $950.
    - For the medium customer truck, the bill will be $350 + $50 x 6 = $300, totaling $650.
    - For the large customer truck, the bill will be $350 + $50 x 2 = $100, totaling $450.
    - Combining all three trucks, the total cost will be $2,050.

### Parameters:
- $D$: Demand or annual consumption
- $S$: Cost of placing an order (ordering cost)
- $H$: Holding cost per unit per year = 25% of USD 1 = USD 0.25/pound

Given:
- Total Annual Demand (All Customers) = 792,000 pounds
- Truck Capacity = 40,000 pounds

### Option 1: Consolidating All Customers into One Truck
For this option:
- $D$ = 792,000 pounds
- $S$ = USD 1,350 (cost for one truck serving all 20 customers)

1. Compute EOQ:
$$ EOQ = \sqrt{\frac{2 \times 792,000 \times 1,350}{0.25}} = \sqrt{\frac{2,136,800,000}{0.25}} \approx 92,572 \text{ pounds} $$

This suggests ordering approximately 92,572 pounds each time. However, given the truck's capacity, DowChem will ship in batches of 40,000 pounds, which is the closest feasible quantity to EOQ.

2. Compute TAC: We use 40,000 pounds as the order quantity as it is the closest feasible quantity to EOQ.
$$ TAC = \frac{D}{EOQ} \times S + \frac{EOQ}{2} \times H = \frac{792,000}{40,000} \times 1,350 + \frac{40,000}{2} \times 0.25 \approx USD 26,730 + USD 5,000 = USD 31,730 $$

### Option 2: Grouping Customers into Two Categories
For each group:
- $D$ = 396,000 pounds
- $S$ = USD 850 (cost for one truck serving 10 customers in the group)

1. Compute EOQ for each group:
$$ EOQ = \sqrt{\frac{2 \times 396,000 \times 850}{0.25}} \approx 65,582 \text{ pounds} $$

Again, due to truck capacity, DowChem will ship in batches of 40,000 pounds for each group.

2. Compute TAC for each group, we use 40,000 pounds as the order quantity as it is the closest feasible quantity to EOQ.
$$ TAC = \frac{D}{EOQ} \times S + \frac{EOQ}{2} \times H \approx USD 8,415 + USD 5,000 = USD 13,415 $$

Since there are two groups, the combined TAC is USD 13,415 x 2 = USD 26,830.

### Option 3: Aggregating Each Customer Type into One Truck
This requires three separate EOQ calculations:

1. **For Small Customers**:
- $D$ = 144,000 pounds
- $S$ = USD 950

$$ EOQ = \sqrt{\frac{2 \times 144,000 \times 950}{0.25}} \approx 48,087 \text{ pounds} $$

TAC for small customers, we use 40,000 pounds as the order quantity as it is the closest feasible quantity to EOQ:
$$ TAC = \frac{D}{EOQ} \times S + \frac{EOQ}{2} \times H \approx USD 3,420 + USD 5,000 = USD 8,420 $$

2. **For Medium Customers**:
- $D$ = 360,000 pounds
- $S$ = USD 650

$$ EOQ = \sqrt{\frac{2 \times 360,000 \times 650}{0.25}} \approx 65,376 \text{ pounds} $$

TAC for medium customers, we use 40,000 pounds as the order quantity as it is the closest feasible quantity to EOQ:
$$ TAC = \frac{D}{EOQ} \times S + \frac{EOQ}{2} \times H \approx USD 5,850 + USD 5,000 = USD 10,850 $$

3. **For Large Customers**:
- $D$ = 288,000 pounds
- $S$ = USD 450

$$ EOQ = \sqrt{\frac{2 \times 288,000 \times 450}{0.25}} \approx 60,415 \text{ pounds} $$

TAC for large customers, we use 40,000 pounds as the order quantity as it is the closest feasible quantity to EOQ:
$$ TAC = \frac{D}{EOQ} \times S + \frac{EOQ}{2} \times H \approx USD 3,240 + USD 5,000 = USD 8,240 $$

Combined TAC for Option 3: USD 8,420 + USD 10,850 + USD 8,240 = USD 27,510

### Recommendation:
Based on the optimal annual total costs computed:
- Option 1: USD 31,730
- Option 2: USD 26,830
- Option 3: USD 27,510

**Option 2** is the best option for DowChem as it has the lowest total annual cost.

Cute-lights Manufacturing Company makes flashing lights for toys. The company operates its production facility 300 days per year. It has orders for about 12,000 flashing lights per year and has the capability of producing 100 per day. The setup cost for each production run is $50. The cost of each light is $1. The holding cost is $0.10 per light per year. 

a) (5 pts) What is the optimal production quantity?

b) (5 pts) What is the annual cost based on the optimal production quantity?

c) (5 pts) In each production cycle, how many days would production continue?

d) (5 pts) In each production cycle, how many lights would be in inventory when production stops?


**a) What is the optimal production quantity?**
To find the optimal production quantity (EPL - Economic Production Lot), we can use the EPL formula:

$$
EPL = \sqrt{\frac{2k\lambda}{h (1 - \frac{\lambda}{P})}}
$$

Where:

- $\lambda$ is the annual demand, which is 12,000 flashing lights.
- $k$ is the setup cost for each production run, which is USD 50.
- $h$ is the holding cost per light per year, which is USD 0.10.
- $P$ is the production rate year, which is 100 flashing lights a day, or 30000 flashing lights a year.

Substituting these values into the formula:

$$
EOQ = \sqrt{\frac{2 \times 50 \times 12,000}{0.10 \times (1 - \frac{12,000}{30,000})}} \approx 4,472
$$

Therefore, the optimal production quantity is 4,472 flashing lights.

**b) What is the annual cost based on the optimal production quantity?**
To calculate the annual cost, we need to consider the following costs:

The Formula is

$$
W(Q) = \frac{k \times \lambda}{Q} + \frac{h \times (1 - \frac{\lambda}{p}) \times Q}{2} + \lambda \times c
$$

Replacing the values:

$$
W(Q) = \frac{50 \times 12,000}{4,472} + \frac{0.10 \times (1 - \frac{12,000}{100 \times 300}) \times 4,472}{2} + 12,000 \times 1
$$

$$
W(Q) = 134.5 + 1,341.6 + 12,000 = 12,268.3
$$

Therefore, the annual cost based on the optimal production quantity is USD 13,476.1.

**c) In each production cycle, how many days would production continue?**
The number of days production would continue is given by:

$$
T = \frac{Q}{p} = \frac{4,472}{100} = 44.72 \text{ days} \approx 45 \text{ days}
$$

Therefore, production would continue for 45 days.

**d) In each production cycle, how many lights would be in inventory when production stops?**
The number of lights in inventory when production stops is given by:

The inventory level at the end of the production cycle is given by:

$$
I = Q \times (1 - \frac{\lambda}{p}) = 4,472 \times (1 - \frac{12,000}{100 \times 300}) = 2683 \text{ lights}
$$

A purchasing agent for a particular type of silicon wafer used in the production of semiconductors must decide among three sources. Source A will sell the silicon wafers for USD 2.50 per wafer, independently of the number of wafers ordered. # Solution

Given:
- $ k $ (order setup cost) = USD 100
- $ \lambda $ (annual demand) = 20,000 wafers
- Annual interest rate = 20%

To find the holding cost per unit ($ h $), we'll compute it as a percentage of the unit cost. But since the unit cost differs across the sources, $ h $ will also differ.

### a) Source Selection & Order Size:

For each source, let's compute the Economic Production Lot (EPL) using the given formula:

$$ EPL = \sqrt{\frac{2k\lambda}{h (1 - \frac{\lambda}{P})}} $$

However, since $ P $ (production rate) is not given, and the context is purchasing rather than production, we can use the EOQ (Economic Order Quantity) model instead:

$$ EOQ = \sqrt{\frac{2k\lambda}{h}} $$

**Source A**:
- Unit cost = USD 2.50
- $ h = 0.20 \times 2.50 = USD 0.50 $
$$ EOQ_A = \sqrt{\frac{2(100)(20000)}{0.50}} $$
= 2828 wafers.

**Source B**:
- Unit cost = USD 2.40
- $ h = 0.20 \times 2.40 = USD 0.48 $
$$ EOQ_B = \sqrt{\frac{2(100)(20000)}{0.48}} $$
= 2887 wafers (but minimum order is 3,000 wafers).

**Source C**:

- $ h = 0.20 \times 2.30 = USD 0.46 $
$$ EOQ_C = \sqrt{\frac{2(100)(20000)}{0.46}} $$
= 2949 wafers (but minimum order is 4,000 wafers).

Now, let's compute the total annual cost for each source:
**Source A**:
$$ TAC_A = \frac{D}{EOQ_A} \times S + \frac{EOQ_A}{2} \times h + D \times c $$
$$ TAC_A = \frac{20000}{2828} \times 100 + \frac{2828}{2} \times 0.50 + 20000 \times 2.50 $$
$$ TAC_A = 707.11 + 707.11 + 50000 = USD 51,414.22 $$
**Source B**:
$$ TAC_B = \frac{D}{EOQ_B} \times S + \frac{EOQ_B}{2} \times h + D \times c $$
$$ TAC_B = \frac{20000}{3000} \times 100 + \frac{3000}{2} \times 0.48 + 20000 \times 2.40 $$
$$ TAC_B = 666.67 + 720 + 48000 = USD 49,386.67 $$
**Source C**:
$$ TAC_C = \frac{D}{EOQ_C} \times S + \frac{EOQ_C}{2} \times h + D \times c $$
$$ TAC_C = \frac{20000}{4000} \times 100 + \frac{4000}{2} \times 0.46 + 20000 \times 2.30 $$
$$ TAC_C = 500 + 920 + 46000 = USD 47,420 $$
Therefore, the optimal source is **Source C** with an order size of 4,000 wafers.

### b) Optimal Holding and Setup Costs given the Optimal Source of 4,000 Wafers:

As Green's has contracted to supply fixed quantities of buttons for the next three years, its demand can be treated as fixed and known. The relevant data for this problem appear in the following table.

| Button | Annual Production Rate | Setup Time (hours) | Variable Type | Sales (units/day) | Cost (usd) |
| ------ | ---------------------- | ------------------- | ------------- | ----------------- | ---------- |
| A      | 25,900                 | 4,500               | 6             | $0.003            |
| B      | 42,000                 | 5,500               | 4             | $0.002            |
| C      | 14,400                 | 3,300               | 8             | $0.008            |

| E      | 12,500                 | 1,800               | 3             | $0.010            |
| F      | 75,000                 | 3,900               | 6             | $0.005            |
| G      | 30,000                 | 2,900               | 1             | $0.004            |
| H      | 18,900                 | 1,200               | 3             | $0.007            |

Assume 250 working days per year. Green's accounting department has established an 18 percent annual interest rate for the cost of capital and a 3 percent annual interest rate to account for storage space. Setup costs are $20 per hour required to reconfigure the equipment for a new style. Suppose that the firm decides to use a rotation cycle policy for production of the buttons.

a) (8 pts) What is the optimal rotation cycle time? 

b) (7 pts) How large should the lots be?  

c) (7 pts) What is the average annual cost of holding and setups at the optimal solution?

More specifically, if Green's agreed to make three shipments per year for each button style, what production policy would you recommend?


```python
import pandas as pd

# Load the data in a dataframe manually

data = {'Button': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
        'Annual Production Rate': [25900, 42000, 14400, 46000, 12500, 75000, 30000, 18900],
        'Setup Time (hours)': [4500, 5500, 3300, 3200, 1800, 3900, 2900, 1200],
        'Variable Type': [6, 4, 8, 4, 3, 6, 1, 3],
        'Sales (units/day)': [0.003, 0.002, 0.008, 0.002, 0.01, 0.005, 0.004, 0.007],
        'Cost (usd)': [0,0,0,0,0,0,0,0]}

df = pd.DataFrame(data)

df
```

```python

# Add the cost column
df['Cost (usd)'] = df['Variable Type'] * df['Sales (units/day)'] * 250

# Add the cost of capital column
df['Cost of Capital'] = df['Annual Production Rate'] * 0.18 * 0.03

# Add the total cost column
df['Total Cost'] = df['Cost (usd)'] + df['Cost of Capital']

# Add the cost of setup column
df['Cost of Setup'] = df['Setup Time (hours)'] * 20

# Add the total cost column
df['Total Cost'] = df['Cost (usd)'] + df['Cost of Capital'] + df['Cost of Setup']

# Add the cost of holding column
df['Cost of Holding'] = df['Annual Production Rate'] * 0.03 * 0.18

# Add the total cost column
df['Total Cost'] = df['Cost (usd)'] + df['Cost of Capital'] + df['Cost of Setup'] + df['Cost of Holding']

df.head(8)
```

```python
# Calculate the optimal rotation cycle time
df['Optimal Rotation Cycle Time'] = df['Variable Type'] * df['Setup Time (hours)'] / df['Annual Production Rate']

# Print the optimal rotation cycle time
print('The optimal rotation cycle time is', df['Optimal Rotation Cycle Time'].max(), 'hours.')
```

```python
# Calculate the optimal lot size
df['Optimal Lot Size'] = df['Annual Production Rate'] / df['Variable Type']

# Print the optimal lot size
print('The optimal lot size is', df['Optimal Lot Size'].max(), 'units.')
```

```python
# Calculate the average annual cost of holding and setups at the optimal solution
df['Average Annual Cost of Holding and Setups at the Optimal Solution'] = df['Cost of Holding'] + df['Cost of Setup']

# Print the average annual cost of holding and setups at the optimal solution
print('The average annual cost of holding and setups at the optimal solution is', df['Average Annual Cost of Holding and Setups at the Optimal Solution'].max(), 'usd.')
```
