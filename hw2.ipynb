# 1.
DowChem, a manufacturer specializing in chemical products, currently serves 20 customers within the same region through its distribution center. These customers can be categorized into three types: small, medium, and large, each with their respective annual consumption, as detailed in the following table:

| Customer Type | Number of Customers | Consumption (Pounds per year) |
|---------------|---------------------|-----------------------------|
| Small         | 12                  | 12,000                      |
| Medium        | 6                   | 60,000                      |
| Large         | 2                   | 144,000                     |

DowChem collaborates with Superior Trucking Co. for all its shipping needs. Each truck employed for transportation has a maximum capacity of 40,000 pounds. Superior charges a fixed fee of $400 for each shipment, irrespective of the quantity transported on the truckâ€”from the distribution center to each customer.

DowChem's current approach is to dispatch a full truckload to each customer as required. Nevertheless, John Smith, the Vice President of DowChem, is deeply concerned about the escalating distribution costs. John approached Superior to inquire about the possibility of consolidating shipments for multiple customers onto a single truck. Superior conveyed that they would charge $350 per truck and an additional $50 for each drop-off they manage. Consequently, if a single truck is used for a solitary delivery, the total charge would amount to $400. However, if a truck has to make four deliveries, the total charge would rise to $550.

The cost incurred by DowChem for each pound of chemical is $1, and the company also bears a holding cost of 25%. John is eager to evaluate several distribution options to determine an improved strategy:

**Option 1:** Consolidating all 20 customers into one truck. Whenever a truck is dispatched, it will serve all customers.

**Option 2:** Grouping the 20 customers into two categories, with one large, three medium, and six small customers in each group. The shipment for each group will be aggregated into a single truck.

**Option 3:** Aggregating each customer type into one truck. This means that large customers are bundled together, and their shipment is consolidated into a single truck, and the same applies to medium and small customers.

**a) (10 pts)** What is the annual cost associated with DowChem's current strategy of delivering full truckloads to each customer to replenish inventory?

**b) (20 pts)** Assuming the computation of the optimal shipment quantity using the EOQ (Economic Order Quantity) model and the optimal annual total cost for each of the delivery options 1, 2, and 3, which delivery option would you recommend for DowChem, and why?


# Solution

### Part a)

To calculate the annual cost of DowChem's current strategy, we need to consider the following costs:

1. **Shipping Cost:** This includes the fixed cost of USD400 per shipment.

2. **Holding Cost:** DowChem incurs a holding cost of 25% for each pound of chemical. We'll calculate this as a percentage of the annual consumption for each customer type and then sum them up.

3. **Total Cost:** The sum of the shipping cost and the holding cost.

Let's calculate these costs:

1. Shipping Cost:

   DowChem has 20 customers, and each customer is served individually. Therefore, the total annual shipping cost can be calculated as follows:

   $$
   \text{Total Shipping Cost} = \text{Number of Customers} \times \text{Cost per Shipment}
   $$

   $$
   \text{Total Shipping Cost} = 20 \times USD400 = USD8,000
   $$

2. Holding Cost:

   For each customer type (small, medium, and large), we calculate the annual holding cost based on their respective consumption. Then, we sum these costs.

   - For small customers:
   
     $$
     \text{Holding Cost for Small Customers} = \text{Annual Consumption} \times \text{Holding Cost Percentage} \times \text{Cost per Pound}
     $$

     $$
     \text{Holding Cost for Small Customers} = 12,000 \text{ pounds} \times 0.25 \times USD1 = USD3,000
     $$

   - For medium customers:

     $$
     \text{Holding Cost for Medium Customers} = 60,000 \text{ pounds} \times 0.25 \times USD1 = USD15,000
     $$

   - For large customers:

     $$
     \text{Holding Cost for Large Customers} = 144,000 \text{ pounds} \times 0.25 \times USD1 = USD36,000
     $$

3. Total Cost:

   Now, we sum up the shipping cost and the holding cost to get the total annual cost:

   $$
   \text{Total Annual Cost} = \text{Total Shipping Cost} + \text{Total Holding Cost}
   $$

   $$
   \text{Total Annual Cost} = USD8,000 + USD3,000 + USD15,000 + USD36,000 = USD62,000
   $$

Therefore, the annual cost of DowChem's current strategy is USD62,000.

### Part b)

To determine the optimal delivery option, we need to calculate the total annual cost for each of the three options: Option 1, Option 2, and Option 3. We'll use the EOQ model to calculate the optimal shipment quantity for each option and then compute the total cost for each.

Let's start with Option 1:

#### Option 1:

In this option, all 20 customers are consolidated into one truck for each shipment.

1. Calculate the EOQ (Economic Order Quantity) for the entire annual demand.

   $$
   EOQ_1 = \sqrt{\frac{2DS}{H}}
   $$

   Where:
   - $D$ = Total annual demand (sum of all customer types)
   - $S$ = Setup (ordering) cost per shipment
   - $H$ = Holding cost per unit per year

   $$
   D = (12 \times 12,000) + (6 \times 60,000) + (2 \times 144,000) = 1,248,000 \text{ pounds}
   $$

   $$
   S = USD400 \text{ per shipment}
   $$

   $$
   H = 0.25 \times USD1 = USD0.25 \text{ per pound per year}
   $$

   $$
   EOQ_1 = \sqrt{\frac{2 \times 1,248,000 \text{ pounds} \times USD400}{0.25 \times USD1}}
   $$

   $$
   EOQ_1 \approx 15811.4 \text{ pounds}
   $$

2. Calculate the number of shipments required:

   $$
   \text{Number of Shipments for Option 1} = \frac{D}{EOQ_1}
   $$

   $$
   \text{Number of Shipments for Option 1} \approx \frac{1,248,000 \text{ pounds}}{15811.4 \text{ pounds}} \approx 79 \text{ shipments}
   $$

3. Calculate the total annual cost for Option 1:

   $$
   \text{Total Annual Cost for Option 1} = (\text{Number of Shipments for Option 1} \times \text{Cost per Shipment}) + (D \times H)
   $$

   $$
   \text{Total Annual Cost for Option 1} = (79 \text{ shipments} \times USD400) + (1,248,000 \text{ pounds} \times USD0.25)
   $$

   $$
   \text{Total Annual Cost for Option 1} \approx USD31,600 + USD312,000 = USD343,600
   $$

Now, let's move on to Option 2:

#### Option 2:

In this option, the 20 customers are divided into two groups, each with its shipment.

1. Calculate the EOQ for each group:

   For the first group (large, three medium, and six small customers):
   
   $$
   D_1 = (1 \times 144,000) + (3 \times 60,000) + (6 \times 12,000) = 420,000 \text{ pounds}
   $$
   
   $$
   EOQ_2 = \sqrt{\frac{2D_1S}{H}} = \sqrt{\frac{2 \times 420,000 \text{ pounds} \times USD400}{0.25 \times USD1}} \approx 17728.9 \text{ pounds}
   $$

   For the second group (same composition):
   
   $$
   D_2 = 420,000 \text{ pounds}
   $$
   
   $$
   EOQ_3 = \sqrt{\frac{2D_2S}{H}} = \sqrt{\frac{2 \times 420,000 \text{ pounds} \times USD400}{0.25 \times USD1}} \approx 17728.9 \text{ pounds}
   $$

2. Calculate the number of shipments required for each group:

   $$
   \text{Number of Shipments for Group 1} = \frac{D_1}{EOQ_2} \approx \frac{420,000 \text{ pounds}}{17728.9 \text{ pounds}} \approx 23.68 \text{ shipments}
   $$

   $$
   \text{Number of Shipments for Group 2} = \frac{D_2}{EOQ_3} \approx \frac{420,000 \text{ pounds}}{17728.9 \text{ pounds}} \approx 23.68 \text{ shipments}
   $$

3. Calculate the total annual cost for Option 2:

   $$
   \text{Total Annual Cost for Option 2} = (\text{Number of Shipments for Group 1} + \text{Number of Shipments for Group 2}) \times \text{Cost per Shipment} + (D_1 + D_2) \times H
   $$

   $$
   \text{Total Annual Cost for Option 2} \approx (23.68 \text{ shipments} + 23.68 \text{ shipments}) \times USD400 + (420,000 \text{ pounds} + 420,000 \text{ pounds}) \times USD0.25
   $$

   $$
   \text{Total Annual Cost for Option 2} \approx 47.36 \text{ shipments} \times USD400 + 840,000 \text{ pounds} \times USD0.25
   $$

   $$
   \text{Total Annual Cost for Option 2} \approx USD18,944 + USD210,000 = USD228,944
   $$

Finally, let's calculate Option 3:

#### Option 3:

In this option, each customer type (small, medium, large) is consolidated into one truck.

1. Calculate the EOQ for each customer type:

   - For small customers:
   
     $$
     D_3 = 12,000 \text{ pounds}
     $$

     $$
     EOQ_4 = \sqrt{\frac{2D_3S}{H}} = \sqrt{\frac{2 \times 12,000 \text{ pounds} \times USD400}{0.25 \times USD1}} \approx 1591.55 \text{ pounds}
     $$

   - For medium customers:

     $$
     D_4 = 60,000 \text{ pounds}
     $$

     $$
     EOQ_5 = \sqrt{\frac{2D_4S}{H}} = \sqrt{\frac{2 \times 60,000 \text{ pounds} \times USD400}{0.25 \times USD1}} \approx 7957.74 \text{ pounds}
     $$

   - For large customers:

     $$
     D_5 = 144,000 \text{ pounds}
     ]

     $$
     EOQ_6 = \sqrt{\frac{2D_5S}{H}} = \sqrt{\frac{2 \times 144,000 \text{ pounds} \times USD400}{0.25 \times USD1}} \approx 19042.97 \text{ pounds}
     ]

2. Calculate the number of shipments required for each customer type:

   - For small customers:
   
     $$
     \text{Number of Shipments for Small Customers} = \frac{D_3}{EOQ_4} \approx \frac{12,000 \text{ pounds}}{1591.55 \text{ pounds}} \approx 7.54 \text{ shipments}
     ]

   - For medium customers:

     $$
     \text{Number of Shipments for Medium Customers} = \frac{D_4}{EOQ_5} \approx \frac{60,000 \text{ pounds}}{7957.74 \text{ pounds}} \approx 7.54 \text{ shipments}
     ]

   - For large customers:

     $$
     \text{Number of Shipments for Large Customers} = \frac{D_5}{EOQ_6} \approx \frac{144,000 \text{ pounds}}{19042.97 \text{ pounds}} \approx 7.54 \text{ shipments}
     ]

3. Calculate the total annual cost for Option 3:

   $$
   \text{Total Annual Cost for Option 3} = (\text{Number of Shipments for Small Customers} + \text{Number of Shipments for Medium Customers} + \text{Number of Shipments for Large Customers}) \times \text{Cost per Shipment} + (D_3 + D_4 + D_5) \times H
   ]

   $$
   \text{Total Annual Cost for Option 3} \approx (7.54 \text{ shipments} + 7.54 \text{ shipments} + 7.54 \text{ shipments}) \times USD400 + (12,000 \text{ pounds} + 60,000 \text{ pounds} + 144,000 \text{ pounds}) \times USD0.25
   ]

   $$
   \text{Total Annual Cost for Option 3} \approx 22.62 \text{ shipments} \times USD400 + 216,000 \text{ pounds} \times USD0.25
   ]

   $$
   \text{Total Annual Cost for Option 3} \approx USD9,048 + USD54,000 = USD63,048
   ]

Now, let's summarize the results:

- Total Annual Cost for Option 1: USD343,600
- Total Annual Cost for Option 2: USD228,944
- Total Annual Cost for Option 3: USD63,048

The most cost-effective delivery option for DowChem is Option 3, with a total annual cost of USD63,048. This option involves aggregating each customer type into one truck, resulting in the lowest distribution cost. Therefore, I recommend DowChem to adopt Option 3 as its distribution strategy.

# 3.
A purchasing agent for a particular type of silicon wafer used in the production of semiconductors must decide among three sources. Source A will sell the silicon wafers for USD 2.50 per wafer, independently of the number of wafers ordered. Source B will sell the wafers for USD 2.40 each but will not consider an order for fewer than 3,000 wafers, and Source C will sell the wafers for USD 2.30 each but will not accept an order for fewer than 4,000 wafers. Assume an order setup cost of USD 100 and an annual requirement of 20,000 wafers. Assume a 20 percent annual interest rate for holding cost calculations.

a) (5 pts) Which source should be used, and what is the size of the standing order?

b) (5 pts) What is the optimal value of the holding and setup costs for wafers when the optimal source is used?

c) (5 pts) If the replenishment lead time for wafers is three months, determine the reorder point based on the on-hand level of inventory of wafers.

d) (10 pts) Assume that two years have passed, and the purchasing agent must recompute the optimal number of wafers to purchase and from which source to purchase them. Source B has decided to accept any size offer but sells the wafers for USD 2.55 each for orders of up to 3,000 wafers and USD 2.25 each for the incremental amount ordered over 3,000 wafers. Source A still has the same price schedule, and Source C went out of business. Now which source should be used?


# Solution

Certainly, I'll help you solve the problems step by step:

**a) Which source should be used, and what is the size of the standing order?**

To determine the optimal source and the size of the standing order, we can use the Economic Order Quantity (EOQ) model.

The EOQ formula is:

$$EOQ = \sqrt{\frac{2DS}{H}}$$

Where:
- $D$ is the annual demand, which is 20,000 wafers.
- $S$ is the setup cost for each order, which is USD 100.
- $H$ is the holding cost per wafer per year, calculated as 20% of the cost of the wafer.

For Source A:
- Cost per wafer ($C_A$) = USD 2.50

For Source B:
- Cost per wafer ($C_B$) = USD 2.40
- Minimum order quantity ($Q_B$) = 3,000 wafers

For Source C:
- Cost per wafer ($C_C$) = USD 2.30
- Minimum order quantity ($Q_C$) = 4,000 wafers

Let's calculate the holding cost ($H$) first:

$$H = 0.20 \times C_A = 0.20 \times USD 2.50 = USD 0.50$$

Now, we can calculate the EOQ for each source:

For Source A:
$$EOQ_A = \sqrt{\frac{2 \times 20,000 \times USD 100}{USD 0.50}} = 4000$$

For Source B (considering $Q_B$):
$$EOQ_B = \sqrt{\frac{2 \times 20,000 \times USD 100}{USD 0.50}} = 4000$$

For Source C (considering $Q_C$):
$$EOQ_C = \sqrt{\frac{2 \times 20,000 \times USD 100}{USD 0.50}} = 4000$$

All three sources have the same EOQ of 4,000 wafers. Therefore, the optimal source to use is the one with the lowest cost per wafer, which is Source C at USD 2.30 per wafer. The size of the standing order is 4,000 wafers.

**b) What is the optimal value of the holding and setup costs for wafers when the optimal source is used?**

To calculate the optimal value of the holding cost ($H$) and the setup cost ($S$) when the optimal source (Source C) is used, we can use the following formulas:

$$H = \frac{EOQ \times H}{2}$$
$$S = \frac{D}{EOQ} \times S$$

Substituting the values:

$$H = \frac{4,000 \times USD 0.50}{2} = USD 2,000$$
$$S = \frac{20,000}{4,000} \times USD 100 = USD 500$$

So, the optimal holding cost for wafers is USD 2,000, and the setup cost is USD 500 when Source C is used.

**c) If the replenishment lead time for wafers is three months, determine the reorder point based on the on-hand level of inventory of wafers.**

The reorder point can be calculated using the following formula:

$$Reorder\ Point = Lead\ Time \times Demand\ Rate$$

Where:
- Lead Time is 3 months, which is equivalent to 3/12 years.
- Demand Rate is the annual demand, which is 20,000 wafers.

$$Reorder\ Point = \frac{3}{12} \times 20,000 = 5,000$$

The reorder point is 5,000 wafers.

**d) Now which source should be used?**

Considering the new pricing for Source B:
- Cost per wafer ($C_B$) for up to 3,000 wafers = USD 2.55
- Cost per wafer ($C_B$) for incremental amount ordered over 3,000 wafers = USD 2.25

Let's recalculate the EOQ for Source B:

$$EOQ_B = \sqrt{\frac{2 \times 20,000 \times USD 100}{USD 0.50}} = 4000$$

Since Source B now accepts any size offer, we can consider both the cost for up to 3,000 wafers and the incremental cost.

For Source B (up to 3,000 wafers):
$$EOQ_B = 4000$$
Total Cost ($TC_B$) for up to 3,000 wafers = $C_B \times 3000 + \frac{2EOQ_B}{H} \times H$
$$TC_B = (USD 2.55 \times 3000) + \frac{2 \times 4000}{USD 0.50} \times USD 0.50$$
$$TC_B = USD 7,650 + USD 4,000 = USD 11,650$$

For Source B (incremental amount over 3,000 wafers):
Total Cost ($TC_B$) for incremental amount = $C_B \times (EOQ - 3000) + \frac{2(EOQ - 3000)}{H} \times H$
$$TC_B = (USD 2.25 \times (4000 - 3000)) + \frac{2(4000 - 3000)}{USD 0.50} \times USD 0.50$$
$$TC_B = USD 2250 + USD 2000 = USD 4250$$

Now, let's compare the total costs:

- Total Cost for Source A = USD 2.30 \times 4000 = USD 9,200
- Total Cost for Source C (previously calculated) = USD 2.30 \times 4000 = USD 9,200
- Total Cost for Source B (up to 3,000 wafers) = USD 11,650
- Total Cost for Source B (incremental amount) = USD 4,250

The lowest total cost is USD 4,250 for Source B (incremental amount). Therefore, Source B should be used now, and the optimal source has changed.

# Solution

a) What is the optimal rotation cycle time?

The optimal rotation cycle time can be determined by using the Economic Order Quantity (EOQ) formula. The EOQ formula is given by:

$$
EOQ = \sqrt{\frac{2DS}{H}}
$$

Where:
- $D$ = Annual demand (the sum of annual sales for all button styles)
- $S$ = Setup cost per order
- $H$ = Holding cost per unit per year

First, we need to calculate the values of $D$, $S$, and $H$.

$D$ (Annual Demand) = Sum of annual sales for all button styles = $\sum$ (Annual Production Rate * Sales)

$$
D = (25,900 * 6) + (42,000 * 4) + (14,400 * 8) + (46,000 * 4) + (12,500 * 3) + (75,000 * 6) + (30,000 * 1) + (18,900 * 3)
$$

Now, let's calculate $S$ (Setup cost per order). We know the setup cost is $20 per hour, and we need to consider the setup time for each button style.

$$
S = 20 \times \sum (Setup Time)
$$

$$
S = 20 \times (4,500 + 5,500 + 3,300 + 3,200 + 1,800 + 3,900 + 2,900 + 1,200)
$$

Next, we calculate $H$ (Holding cost per unit per year). The formula for $H$ is given as:

$$
H = IC \times r
$$

Where:
- $IC$ = Inventory carrying cost per unit per year
- $r$ = Annual interest rate for storage space

$IC$ can be calculated as:

$$
IC = \frac{C}{2}
$$

Where $C$ is the cost per unit, which is the variable cost (Cost column in the table).

$$
IC = \frac{0.003 + 0.002 + 0.008 + 0.002 + 0.010 + 0.005 + 0.004 + 0.007}{2}
$$

Now, we can calculate $H$ using the given annual interest rate for storage space ($r$) of 3%.

$$
H = IC \times r = \left(\frac{0.003 + 0.002 + 0.008 + 0.002 + 0.010 + 0.005 + 0.004 + 0.007}{2}\right) \times 0.03
$$

Now that we have calculated $D$, $S$, and $H$, we can plug these values into the EOQ formula to find the optimal rotation cycle time.

$$
EOQ = \sqrt{\frac{2DS}{H}}
$$
